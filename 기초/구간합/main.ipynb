```python
# 위 문제 투포인터로 풀기
n = int(input("수 입력하세요: "))

count = 1
start_idx = 1
end_idx = 1

sum = 1

while end_idx != n:
    if sum == n:
        count += 1
        end_idx += 1
        sum += end_idx
    elif sum < n:
        end_idx += 1
        sum += end_idx
    else:
        sum -= start_idx
        start_idx += 1
print(count)
```

N개의 수 A1, A2, ..., AN과 L이 주어진다. Ai-L+1 ~ Ai 중 최솟값을 Di라고 할 때 D에 저장된 수를 출력하는 프로그램을 작성하시오. 이때 i ≤ 0 인 Ai는 무시하고 D를 구해야 한다.

```python
L = int(input("L 입력하세요: "))

numbers = list(map(int, input("수열 입력하세요: ").split()))

D = [0] * len(numbers)

for i in range(len(numbers)):
    start = 0
    if i < L:
        ## i = 2 이면  
        start = 0
    else:
        start = i - L + 1
    temp_list = numbers[start:i+1]
    D[i] = min(temp_list)

print(D)
```

```python
### 큐 이용
queue = [] 
D = [0] * len(numbers)
L = int(input("L 입력하세요: "))

numbers = list(map(int, input("수열 입력하세요: ").split()))

for i in range(len(numbers)):
    queue.append(numbers[i])
    if len(queue) < L:
        D[i] = min(queue)
    else:
        D[i] = min(queue[i-L+1:i+1])
print(D)
```

```python
### 본격적 queue 이용
from collections import deque
N,L = map(int, input("N, L 입력하세요: ").split())
mydeque = deque() #(value, index)
now = list(map(int, input("수열 입력하세요: ").split()))
for i in range(N):
    while mydeque and mydeque[-1][0] > now[i]: ## 큐 마지막 값이, 들어올 now[i]보다 크면 얘는 절대로 최소값이 될 수 없다.
        ## 위 while절이 핵심임. 이걸 여러번 하니까, 결과적으로 
        ## 들어올 now[i]보다는 작은게 왼쪽에 있을거임.
        mydeque.pop() ## 큐에서 제거
    mydeque.append((now[i], i)) ## 큐에 (값, 인덱) 추가
    ## q[i-L+1] ~ q[i] 범위에서 최소값이 필요하다.
    if mydeque[0][1] <= i - L: ## 인덱스가 i-L+1보다 작으면 가치가 없음
        mydeque.popleft() ## 큐에서 제거
    print(mydeque[0][0], end=' ') ## 큐의 첫번째 값이 최소값

```
