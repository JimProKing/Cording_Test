```python
### 연속된 자릿수 합으로 나타낼 수 있는 가짓수 출력

def function(x):
    cnt = 0
    for i in range(x):
        for j in range(i,x):
            if sum(range(i,j+1)) == x:
                cnt += 1
    return cnt

temp = int(input("수 입력하세요: "))
print(function(temp))
```

```python
# 위 문제 투포인터로 풀기
n = int(input("수 입력하세요: "))

count = 1
start_idx = 1
end_idx = 1

sum = 1

while end_idx != n:
    if sum == n:
        count += 1
        end_idx += 1
        sum += end_idx
    elif sum < n:
        end_idx += 1
        sum += end_idx
    else:
        sum -= start_idx
        start_idx += 1
print(count)
```
