````python
## 버블정렬 O(n^2)

n = int(input("수의 개수를 입력하세요: "))
temp = []
for i in range(n):
    temp.append(int(input("수열에 들어갈 수를 입력하세요: ")))

for i in range(n):
    for j in range(0, n-i-1):
        if temp[j] > temp[j+1]:
            temp[j], temp[j+1] = temp[j+1], temp[j]
print(temp)
````

````python
# 선택정렬 O(n^2)
n = int(input("수 입력"))
temp = list(str(n))
for i in range(len(temp)):
    x = i
    for j in range(i+1, len(temp)):
        if temp[j] > temp[x]:
            x = j
    temp[i], temp[x] = temp[x], temp[i]
print("".join(temp))
````

````python
# 삽입정렬 O(n^2)
n = int(input("수 입력"))
temp = list(str(n))          # temp가 이미 오름차순이라고 가정
new_num = int(input("새로운 수 입력"))

for i in range(len(temp)):
    if new_num <= int(temp[i]):   # ✅ if만 변경(핵심)
        temp.insert(i, str(new_num))
        break
else:
    temp.append(str(new_num))

print("".join(temp))
````

````python
## 퀵정렬 O(nlogn) 최악 O(n^2)
def swap(A,i,j):
    temp = A[i]
    A[i] = A[j]
    A[j] = temp

def quick_sort(A,s,e):
    if s>=e:
        return A
    swap(A,s,e) ## 지금 A[s]에 있는게 피벗
    i, j = s+1, e
    while True:
        while i <= e and A[s] > A[i]:
            i += 1
        while j > s and A[s] < A[j]: ## s==j이면 나가야 함 (안전장치) 그래서 초과조건
            j -= 1
        if i >= j:
            break
        swap(A,i,j)
    ## while문 나왔다는건, i==j가 되었다는 이야기
    ## 해야 할 것 - 피벗 위치(s)와 현재 위치(i)바꾸고 재귀 수행
    ## i==j이면 상관없지만, 엇갈리는 경우 발생 가능해서 j기준으로
    swap(A,s,j)
    quick_sort(A,s,j-1)
    quick_sort(A,j+1,e)
    return A

A = list(map(int, input("데이터 입력").split()))
print(quick_sort(A,0,len(A)-1))
````

````python
# 병합정렬 O(nlogn)
def merge_sort(A):
    if len(A) <= 1:
        return A
    mid = len(A) // 2
    left = merge_sort(A[:mid])
    right = merge_sort(A[mid:])
    merged = []
    i, j = 0, 0
    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            merged.append(left[i])
            i += 1
        else:
            merged.append(right[j])
            j += 1
    merged.extend(left[i:])
    merged.extend(right[j:])
    return merged
````

````python
# 병합정렬 O(nlogn)
def merge(left, right):
    merged = []
    i = j = 0
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            merged.append(left[i]); i += 1
        else:
            merged.append(right[j]); j += 1
    ## 지금 뭔가 남아있더라도, left[i]와 right[j]는 이미 정렬되어 있기 때문에, 그냥 뒤에 붙여주면 됨
    merged += left[i:]
    merged += right[j:]
    return merged

def merge_sort(A):
    if len(A) <= 1:
        return A
    mid = len(A) // 2
    return merge(merge_sort(A[:mid]), merge_sort(A[mid:]))

# left  = [1, 4, 7]
# right = [2, 3, 9]

temp = list(map(int, input("데이터 입력").split()))
print(merge_sort(temp))
````
