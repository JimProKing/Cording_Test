````python
## 삽입정렬 (같은 값 있으면 순서유지)
def insertion_sort_stable(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and arr[j] > key: ## 키보다 큰 것만 하나씩 밀고, 가존 위치에 key 삽입  
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
    return arr

arr = [5, 2, 9, 1, 5, 6]
sorted_arr = insertion_sort_stable(arr)
print(sorted_arr)
````

````python
## 퀵정렬 O(nlogn) 최악 O(n^2)
def swap(A,i,j):
    temp = A[i]
    A[i] = A[j]
    A[j] = temp

def quick_sort(A,s,e):
    if s>=e:
        return A
    swap(A,s,e) ## 지금 A[s]에 있는게 피벗
    i, j = s+1, e
    while True:
        while i <= e and A[s] > A[i]:
            i += 1
        while j > s and A[s] < A[j]: ## s==j이면 나가야 함 (안전장치) 그래서 초과조건
            j -= 1
        if i >= j:
            break
        swap(A,i,j)
    ## while문 나왔다는건, i==j가 되었다는 이야기
    ## 해야 할 것 - 피벗 위치(s)와 현재 위치(i)바꾸고 재귀 수행
    ## i==j이면 상관없지만, 엇갈리는 경우 발생 가능해서 j기준으로
    swap(A,s,j)
    quick_sort(A,s,j-1)
    quick_sort(A,j+1,e)
    return A

A = list(map(int, input("데이터 입력").split()))
print(quick_sort(A,0,len(A)-1))
````

````python
# 병합정렬 O(nlogn)
def merge_sort(A):
    if len(A) <= 1:
        return A
    mid = len(A) // 2
    left = merge_sort(A[:mid])
    right = merge_sort(A[mid:])
    merged = []
    i, j = 0, 0
    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            merged.append(left[i])
            i += 1
        else:
            merged.append(right[j])
            j += 1
    merged.extend(left[i:])
    merged.extend(right[j:])
    return merged
````

````python
# 병합정렬 O(nlogn)
def merge(left, right):
    merged = []
    i = j = 0
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            merged.append(left[i]); i += 1
        else:
            merged.append(right[j]); j += 1
    ## 지금 뭔가 남아있더라도, left[i]와 right[j]는 이미 정렬되어 있기 때문에, 그냥 뒤에 붙여주면 됨
    merged += left[i:]
    merged += right[j:]
    return merged

def merge_sort(A):
    if len(A) <= 1:
        return A
    mid = len(A) // 2
    return merge(merge_sort(A[:mid]), merge_sort(A[mid:]))

# left  = [1, 4, 7]
# right = [2, 3, 9]

temp = list(map(int, input("데이터 입력").split()))
print(merge_sort(temp))
````

````python
from collections import deque
def sliding_sort(arr,size_slide):
    answer = []
    dq = deque()
    for i,x in enumerate(arr):
        while dq and dq[-1][1]>x: ## 진짜 필요없는애들 (어떤 경우에도, x보다 큰 앞에놈은 절대 안쓴다.)
            dq.pop()
        dq.append((i,x))

        while dq and dq[0][0] <= i-size_slide:
            dq.popleft()
        
        if i>=size_slide-1:
            answer.append(dq[0][1])
    return answer

arr = list(map(int, input("데이터 입력").split()))
slide = int(input("슬라이드 크기 입력"))
print("입력된 배열:", arr)
print("슬라이드 정렬 결과:", sliding_sort(arr, slide))
````

```py
def radix_sort(A):
    if not A:
        return A

    max_num = max(A)
    base = 1

    while max_num // base > 0:
        que = [[] for _ in range(10)]   # 매 패스 초기화

        for x in A:
            digit = (x // base) % 10
            que[digit].append(x)        # append는 순서 유지(안정)

        A = []
        for bucket in que:
            A.extend(bucket)            # pop하지 말고 그대로 이어붙이기(안정)

        base *= 10

    return A
``` 
