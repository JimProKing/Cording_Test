```python
## 이 수열을 스택으로 만들 수 있는가?
### 정답 따라쓰기
N = int(input("수 입력하시오"))
A = [0]*N

for i in range(N):
    A[i] = int(input())

stack = []
num = 1
result = True
answer = []

for i in range(N):
    su = A[i]

    if su >= num: ## su랑 같아질때까지 계속 푸시
        ## A[i]랑 같아질때까지 계속 푸시
        while su >= num:
            stack.append(num)
            num+=1
            answer.append('+')
        ## su가 더 작으면 
        stack.pop()
        answer.append('-')
    else: # 지금 pop하면 A[i]
        n = stack.pop() ## 이건 무조건 A[i]
        ## 여기서 판단해. 팝 한게 수보다 더 크면 이건 못만들어.
        if n > su:
            print("NO")
            result = False
            break
        else:
            answer.append('-')
if result:
    for i in answer:
        print(i)

```

```python
N = int(input())
numbers = [int(input()) for _ in range(N)]

stack = []
answer = []
num = 1

for x in numbers:
    # x까지 push (x 포함)
    while num <= x:
        stack.append(num)
        answer.append('+')
        num += 1

    # 이제 top이 x여야 pop 가능
    if stack and stack[-1] == x:
        stack.pop()
        answer.append('-')
    else:
        print("NO")
        break
else:
    print('\n'.join(answer))
```

```python
# 카드 게임

n = int(input("수 입력하시오"))

temp = []
for i in range(1,n+1):
    temp.append(i)

while len(temp) > 1:
    temp.pop(0)
    temp.append(temp.pop(0))

print(temp[0])
```

```python
# 절대 힙
# x가 양수면 x를 힙에 넣는다.
# x가 0이면 힙에서 절댓값이 가장 작은 값을 출력하고, 그 값을 힙에서 제거한다.

n = int(input("연산 횟수 입력하시오: "))
temp = []
for _ in range(n):
    x = int(input())
    if x > 0 or x < 0:
        temp.append(x)
    else:
        if len(temp) == 0:
            print(0)
        else:
            temp.sort(key=lambda x: (abs(x), x))
            ## temp.sort(key=abs) -> 절댓값 기준으로 정렬
            ## 이렇게만 하면, 동률일때 기준이 없음
            print(temp[0])
            temp.pop(0)
```
