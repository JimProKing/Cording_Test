```python
# 정점 3개
N = 3
graph = [[] for _ in range(N+1)] ## 0번은 안쓸려고 함
```

```python
# map(함수, 반복가능객체)
temp = [1,2,3,4,5,6,7,8,9]

b = map(lambda x: x**2, temp)
print(b)
print(list(b))
```

```python
# 정점 수와 간선 수 입력
N,E = map(int, input().split())
## 3 4 이런식으로 입력하면, 노드 3개 간선 4개짜리 그래프


# 간선 정보 입력받아 저장
for _ in range(E):
    s, e, w = map(int, input().split())
    graph[s].append((e,w))
## 1 2 4 이런식으로 입력하면, 1번 노드에서 2번 노드로 가는 간선의 가중치가 4라는 뜻
## graph[x] :: x번 노드에서 시작되는 간선 정보가 저장된 리스트
```

```python
## 1번 노드에서 시작되는 에지데이터 가져오는 코드

for nextNode, weight in graph[1]:
    ## 1번 노드에서 시작되는 에지 데이터 가져오는 코드
    print(nextNode, weight)
```
