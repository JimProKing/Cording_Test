## 배열: 메모리 연속 공간에 값 有
## 리스트: 배열과 달리 메모리 연속 공간에 값이 저장되지 않음. 대신 각 요소가 다음 요소의 주소를 가리키는 방식으로 구현됨
## 리스트는 배열보다 메모리를 더 많이 사용하지만, 요소의 추가 및 삭제가 용이함. 배열은 고정된 크기를 가지며, 요소의 추가 및 삭제가 어렵지만, 메모리를 효율적으로 사용할 수 있음

## 파이썬에서는 배열과 리스트를 구분하지 않는다.

```python
# 숫자 합 구하기
## N개 숫자 공백없이 - 이 숫자 모두 합해 출력하는 프로그램
n = int(input())
print(n) ## 입력받은 숫자 확인
numbers = list(input())
result = 0

for i in numbers:
    result += int(i)
print(result)


## 오답노트
# n = int(input()) 이렇게만 해도, str 기준 리스트로 저장됨.
```

```python
## 형변환
temp = "32123"

temp_int = int(temp)
print("int",temp_int)

temp_float = float(temp)
print("float",temp_float)

temp_str = str(temp)
print("str",temp_str)

temp_chr = chr(temp_int)
print("chr",temp_chr)

temp_bool = bool(temp)
print("bool",temp_bool)
```

```python
# 새준이의 점수조작
# 조작할 과목의 점수 = (찐 점수 / 최댓값) * 100

#input: 1) 과목 수 2) 각 과목의 점수
n = int(input())
scores = list(map(int, input().split()))

max_score = max(scores)

new_scores = [(score / max_score * 100) for score in scores]

## 새 평균
new_average = sum(new_scores) / n

print(new_average)
```
