```python
## 수학 내림차, 수학 같으면 영어 내림차
scores = [
    {'english': 90, 'math': 80},
    {'english': 80, 'math': 90},
    {'english': 90, 'math': 70},
    {'english': 80, 'math': 80}
]

## A.sort(key=함수)
## lambda 안쓰면 이렇게

# def key_func(x):
#     return (-x['math'], -x['english'])
# scores.sort(key=key_func)

# scores.sort(key=key_func)

scores.sort(key=lambda x: (-x['math'], -x['english']))

for score in scores:
    print(score)
```
