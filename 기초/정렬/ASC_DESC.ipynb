```python
# 오름차 정렬

# 원본 리스트 선언
A = [5,3,2,4,1,]

# 방법1. 리스트 자체를 정렬
A.sort() ## 객체 자체를 정렬하며, 반환값은 None 
print(A)
# 다시 초기화
A = [5,3,2,4,1,]

# 방법2. sorted() 함수를 이용하여 정렬된 새로운 리스트 생성
B = sorted(A)
print(B)
```

```python
# 내림차 정렬

# 원본 리스트 선언
A = [5,3,2,4,1,]
# 방법1. 리스트 자체를 정렬
A.sort(reverse=True) ## 객체 자체를 정렬하며, 반환값은 None
print(A)

# 다시 초기화
A = [5,3,2,4,1,]

# 방법2. sorted() 함수를 이용하여 정렬된 새로운 리스트 생성
B = sorted(A, reverse=True)
print(B)
```

```python
# 내림차 정렬 해야하는데, 직접 제어 못하는 경우
# 음수 이용

A = [5,3,2,4,1,]

A = [-x for x in A] # 음수로 변환
```
