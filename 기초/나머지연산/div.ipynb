```python
## 나머지 분배법칙
print(26%3) ## 결과는 2

## (A+B)%C = (A%C + B%C)%C
print((20%3+6%3)%3) ## 결과는 2

## 뺄셈 분배법칙
## (A-B)%C = (A%C - B%C)%C
print((20-6)%3) ## 결과는 2
print((20%3-6%3)%3) ## 결과는 2
```

```python
# 곱셈의 분배법칙
## (A*B)%C = ((A%C) * (B%C))%C
print((20*6)%3) ## 결과는 0
print(((20%3) * (6%3))%3) ## 결과는
```

```python
# 나눗셈 분배법칙은 성립안함
## (A//B)%C != ((A%C) // (B%C))%C
print((20//6)%3) ## 결과는 0
print(((20%3) // (2%3))%3) ## 결과는 2
```

```python
## 1~10000까지 곱한 값 1000000007로 나눈 나머지 구하기
import time

MOD = 1000000007
result = 1
start = time.time()

for i in range(1, 100001):
    result = (result * i) % MOD ## 어차피 곱셈 분배법칙 성립하니까, mod 연산은 매 반복문마다 수행

end = time.time()
print("result:", result) ## 결과는 457992974
print("Runtime:", end - start)

## 파이썬에서 수가 커지면 연산 속도 느려지니까, 분배법칙 쓸 수 있으면 걍 쓰자
```
